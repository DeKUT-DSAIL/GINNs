```python
import pandas as pd
gompertz_df = pd.read_csv('/kaggle/input/datasets/hypnotu/hust-preprocessed-data/hust_gompertz_params.csv')
gompertz_df.head()
df = gompertz_df.copy()
df['SoH above 80']= df['SoH_above_80'].tolist()
df['normalized SoH']= df['norm_SoH'].tolist()
gompertz_df = df.drop(columns=['SoH_above_80', 'norm_SoH'])
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
# Use a font with better mathematical support
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
# Force the minus sign to be mathematically correct and avoid hyphens
matplotlib.rcParams['axes.unicode_minus'] = False # This forces the unicode minus sign (U+2212)
import numpy as np
```

```python
def gompertz_func(x,k,a,b):
    return k*np.exp(-np.exp(a-(b*x)))
```

```python
gompertz_df['gompertz'] = gompertz_func(x=gompertz_df['rul']/10000,k=gompertz_df['k'],a=gompertz_df['a'],b=gompertz_df['b'])
gompertz_df['gompertz 1st derivative']=gompertz_df['b']*gompertz_func(x=gompertz_df['rul']/10000,k=gompertz_df['k'],a=gompertz_df['a'],b=gompertz_df['b'])*np.exp(gompertz_df['a'] - (gompertz_df['b'] * gompertz_df['rul']/10000))
gompertz_df['gompertz 2nd derivative']=gompertz_df['b']**2*gompertz_func(x=gompertz_df['rul']/10000,k=gompertz_df['k'],a=gompertz_df['a'],b=gompertz_df['b'])*np.exp(gompertz_df['a'] - (gompertz_df['b'] * gompertz_df['rul']/10000))*(np.exp(gompertz_df['a'] - (gompertz_df['b'] * gompertz_df['rul']/10000)-1))
gompertz_df = gompertz_df.sort_values(by='rul').reset_index(drop=True)
```

```python
label_mapping = {
    'k': '$k$',
    'a': '$a$',
    'b': '$b$',
    'rul': 'RUL',
    'SoH': 'SoH',
    'SoH above 80': 'SoH $> 80$',
    'normalized SoH': 'Norm. SoH',
    'gompertz': '$G(x)$',
    'gompertz 1st derivative': "$G'(x)$",
    'gompertz 2nd derivative': "$G''(x)$"
}
# Assuming 'df' is your dataframe
gompertz_df = gompertz_df.rename(columns=label_mapping)

# Compute the correlation matrix
corr_matrix = gompertz_df.iloc[:, 1:].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Analysis of Gompertz Parameters and Battery Metrics")
plt.xticks(rotation=45, ha='right')
plt.show()
```
